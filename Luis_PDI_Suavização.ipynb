```python
import cv2
import numpy as np
import imutils 
from matplotlib import pyplot as plt
from google.colab import files
from google.colab.patches import cv2_imshow
from skimage.util import random_noise
%matplotlib inline
```

```python
uploaded = files.upload()
```

```python
img = cv2.imread('goku.jpg')
RGB_img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
```

## **ADICIONANDO RUIDOS**



```python
I = RGB_img
gauss = random_noise(I, mode='gaussian', seed=None, clip=True)
sp = random_noise(I, mode='s&p', seed=5, clip=True)

fig1 = plt.figure(figsize=(20,50))
ax1 = fig1.add_subplot(131)
plt.imshow(I)
plt.xticks([]), plt.yticks([])
plt.title('Original')

ax2 = fig1.add_subplot(132)
plt.imshow(gauss)
plt.xticks([]), plt.yticks([])
plt.title('Gaussian')

ax3 = fig1.add_subplot(133)
plt.imshow(sp)
plt.xticks([]), plt.yticks([])
plt.title('Salt & Pepper')

plt.show()
```

```python
imgNC = cv2.cvtColor(RGB_img, cv2.COLOR_RGB2GRAY)
cv2.imwrite("Cinza.jpg",imgNC)
gray = cv2.imread("Cinza.jpg")
```

```python
I1 = gray
gauss1 = random_noise(I1, mode='gaussian', seed=None, clip=True)
sp1 = random_noise(I1, mode='s&p', seed=5, clip=True)

fig1 = plt.figure(figsize=(20,50))
ax1 = fig1.add_subplot(131)
plt.imshow(I1)
plt.xticks([]), plt.yticks([])
plt.title('Original')

ax2 = fig1.add_subplot(132)
plt.imshow(gauss1,'gray')
plt.xticks([]), plt.yticks([])
plt.title('Gaussian')

ax3 = fig1.add_subplot(133)
plt.imshow(sp1,'gray')
plt.xticks([]), plt.yticks([])
plt.title('Salt & Pepper')

plt.show()
```

## **Filtro da Média (FUNÇÃO BLUR)**

```python
blur = cv2.blur(I,(10,10))

fig3 = plt.figure(figsize=(15,20))
ax1 = fig3.add_subplot(121)
plt.imshow(I),plt.title('Original')
plt.xticks([]), plt.yticks([])

ax2 = fig3.add_subplot(122)
plt.imshow(blur),plt.title('Blurred')
plt.xticks([]), plt.yticks([])

plt.show()
```

```python
blur1 = cv2.blur(I1,(10,10))

fig4 = plt.figure(figsize=(15,20))
ax1 = fig4.add_subplot(121)
plt.imshow(I1,'gray'),plt.title('gray')
plt.xticks([]), plt.yticks([])

ax2 = fig4.add_subplot(122)
plt.imshow(blur1,'gray'),plt.title('Gray Blurred')
plt.xticks([]), plt.yticks([])

plt.show()
```
